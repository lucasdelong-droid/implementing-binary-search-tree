```python
class TreeNode:
    def __init__(self, student_id, name, left = None, right = None):
        self.student_id = student_id
        self.name = name
        self.left = left
        self.right = right

class BinaryTree:
    def __init__(self):
        self.root = None
        self.values = []
    
    def search(self, node, student_id):
        if node is None:
            return node
        if node.student_id == student_id:
            return node
        
        left_result = self.search(node.left, student_id)

        if left_result:
            return left_result
        
        return self.search(node.right, student_id)
    
    def insert(self, student_id, name):
        if self.root is None:
            self.root = TreeNode(student_id, name)
        else:
            self._insert(self.root, student_id, name)
    
    def _insert(self, node, student_id, name):
        if node.left is None:
            node.left = TreeNode(student_id, name)
        elif node.right is None:
            node.right = TreeNode(student_id, name)
        else:
            self._insert(node.left, student_id, name)

    def update(self, student_id, new_name):
        node = self.search(self.root, student_id)

        if node:
            node.name = new_name

    def delete(self, student_id):
        self.root = self._delete(self.root, student_id)
    
    def _delete(self, node, student_id):
        if node is None:
            return node, None
        
        if node.student_id == student_id:
            deleted = node

            #leaf cases
            if node.left is None and node.right is None:
                return None, deleted
            
            #single child cases
            if node.left is None:
                return node.right, deleted
            if node.right is None:
                return node.left, deleted
            
            #two children cases
            successor = self._find_min(node.right) #minimum value of the right side
            node.student_id = successor.student_id #the inputted node's id is now the minimum right side id
            node.name = successor.name #inputted node's name is minimum right side name
            node.right, _ = self._delete(node.right, successor.student_id) # _ ignores the deleted variable of _delete func

        else:
            node.left, deleted = self._delete(node.left, student_id)

            if deleted is None:
                node.right, deleted = self._delete(node.right, student_id)

        return node, deleted
    
    def _find_min(self, node):
        while node.left is not None:
            node = node.left
        return node
    
    def create_full_binary_tree(self):
        self.values = [(1, "Alice"), (2, "Bob"), (3, "Charlie"),
        (4, "David"), (5, "Eve"), (6, "Frank"), (7, "Grace")] 
        
        nodes = [TreeNode(id, name) for id, name in self.values]

        self.root = nodes[0]
        self.root.left = nodes[1]
        self.root.right = nodes[2]

        nodes[1].left = nodes[3]
        nodes[1].right = nodes[4]

        nodes[2].left = nodes[5]
        nodes[2].right = nodes[6]

    def create_complete_binary_tree(self):
        self.values = [(1, "Alice"), (2, "Bob"), (3, "Charlie"),
        (4, "David"), (5, "Eve"), (6, "Frank"), (7, "Grace")] 

        nodes = [TreeNode(id, name) for id, name in self.values]

        self.root = nodes[0]

        nodes[0].left = nodes[1]
        nodes[0].right = nodes[2]

        nodes[1].left = nodes[3]
        nodes[1].right = nodes[4]

        nodes[2].left = nodes[5]
    
    def create_degenerate_tree(self):
        self.values = [(1, "Alice"), (2, "Bob"), (3, "Charlie"),
        (4, "David"), (5, "Eve"), (6, "Frank"), (7, "Grace")] 

        nodes = [TreeNode(id, name) for id, name in self.values]

        self.root = nodes[0]

        current = self.root
        for node in nodes[1:]:
            current.right = node
            current = node
    
    def create_skewed_tree(self):
        self.values = [(1, "Alice"), (2, "Bob"), (3, "Charlie"),
        (4, "David"), (5, "Eve"), (6, "Frank"), (7, "Grace")] 

        nodes = [TreeNode(id, name) for id, name in self.values]

        self.root = nodes[0]

        current = self.root
        for node in nodes[1:]:
            current.right = node
            current = node

def display_tree():
    print("Select a tree: ")
    print("1. Full Binary Tree")
    print("2. Complete Binary Tree")
    print("3. Degenerate Binary Tree")
    print("4. Skewed Binary Tree")
    print("5. Exit")

def display_operation_menu(tree_name, values):
    print(f"Tree name: {tree_name}")
    print(f"Initial values in tree: {values}")
    print("1. Insert Node")
    print("2. Search for Node")
    print("3. Update Node Name")
    print("4. Delete Node")
    print("5. Back to Tree Selection")

def main():
    dataset = [(1, "Alice"), (2, "Bob"), (3, "Charlie"), (4, "David"), (5, "Eve"), (6, "Frank"), (7, "Grace")]
    print(f"Dataset: {dataset}")

    tree_full = BinaryTree()
    tree_full.create_full_binary_tree()

    tree_complete = BinaryTree()
    tree_complete.create_complete_binary_tree()

    tree_degenerate = BinaryTree()
    tree_degenerate.create_degenerate_tree()

    tree_skewed = BinaryTree()
    tree_skewed.create_skewed_tree()

    trees = {
        '1': (tree_full, "Full Tree"),
        '2': (tree_complete, "Complete Tree"),
        '3': (tree_degenerate, "Degenerate Tree"),
        '4': (tree_skewed, "Skewed Tree")
    }

    current_tree = None

    while True:
        if current_tree is None:
            display_tree()
            select = input("Select a tree")
            if select in trees:
                current_tree, tree_name = trees[select]
            elif select == '5':
                print("Exit")
                break
            else:
                display_operation_menu(tree_name, current_tree.values)
                select = input("Chose an operation")
                if select == 1:
                    student_id = int(input("Enter student id"))
                    name = input("Enter name")
                    current_tree.insert(student_id, name)
                if select == 2:
                    student_id = input("Search for id")
                    current_tree.search(student_id)
                    print("Name was found")
                if select == 3:
                    new_name = input("Enter a new name")
                    current_tree.update(student_id, new_name)
                if select == 4:
                    student_id = input("Choose an id to delete")
                    current_tree.delete(student_id)
                if select == 5:
                    current_tree.display_tree()

main()
```
