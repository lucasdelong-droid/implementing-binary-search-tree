```python
class TreeNode:
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None
        self.values = []
    
    def search(self, node, student_id):
        if node is None or node.student_id == student_id:
            return node
        
        left_result = self.search(node.left, student_id)

        if left_result:
            return left_result
        
        return self.search(node.right, student_id)
    
    def insert(self, student_id, name):
        if self.root is None:
            self.root = TreeNode(student_id, name)
        else:
            self._insert(self.root, student_id, name)
    
    def _insert(self, node, student_id, name):
        if node.left is None:
            node.left = TreeNode(student_id, name)
        elif node.right is None:
            node.right = TreeNode(student_id, name)
        else:
            node._insert(self.left, student_id, name)

    def update(self, student_id, new_name):
        node = self._search(self.root, student_id)

        if node:
            node.name = new_name

    def delete(self, student_id):
        self.root = self._delete(self.root, student_id)
        return self.root
    
    def _delete(self, node, student_id):
        if node is None:
            return node, None
        
        if node.student_id == student_id:
            deleted = node

            #leaf cases
            if node.left is None and node.right is None:
                return None, deleted
            
            #single shild cases
            if node.left is None:
                return node.right, deleted
            if node.right is None:
                return node.left, deleted
            
            #two children cases
            successor = self._find_min(node.right) #minimum value of the right side
            node.student_id = successor.student_id #the inputted node's id is now the minimum right side id
            node.name = successor.name #inputted node's name is minimum right side name
            node.right, _ = self._delete(node.right, successor.student_id) # _ ignores the deleted variable of _delete func

        else:
            node.left, deleted = self._delete(node.left, student_id)

            if deleted is None:
                node.right, deleted = self._delete(node.right, student_id)

        return node, deleted
    
    def _find_min(self, node):
        while node.left is not None:
            node = node.left
        return node
    
    def create_full_binary_tree(self):
        self.values = [(1, "Alice"), (2, "Bob"), (3, "Charlie"),
        (4, "David"), (5, "Eve"), (6, "Frank"), (7, "Grace")] 
        
        nodes = [TreeNode(id, name) for id, name in self.values]

        self.root = nodes[0]
        self.root.left = nodes[1]
        self.root.right = nodes[2]

        nodes[1].left = nodes[3]
        nodes[1].right = nodes[4]

        nodes[2].left = nodes[5]
        nodes[2].right = nodes[6]

    def create_complete_binary_tree(self):
        self.values = [(1, "Alice"), (2, "Bob"), (3, "Charlie"),
        (4, "David"), (5, "Eve"), (6, "Frank"), (7, "Grace")] 

        nodes = [TreeNode(id, name) for id, name in self.values]

        self.root = nodes[0]

        nodes[0].left = nodes[1]
        nodes[0].right = nodes[2]

        nodes[1].left = nodes[3]
        nodes[1].right = nodes[4]

        nodes[2].left = nodes[5]
    
    def create_degenerate_tree(self):
        self.values = [(1, "Alice"), (2, "Bob"), (3, "Charlie"),
        (4, "David"), (5, "Eve"), (6, "Frank"), (7, "Grace")] 

        nodes = [TreeNode(id, name) for id, name in self.values]

        self.root = nodes[0]

        current = self.root
        for node in nodes[1:]:
            current.right = node
            current = node
    
    def create_skewed_tree(self):
        self.values = [(1, "Alice"), (2, "Bob"), (3, "Charlie"),
        (4, "David"), (5, "Eve"), (6, "Frank"), (7, "Grace")] 

        nodes = [TreeNode(id, name) for id, name in self.values]

        self.root = nodes[0]

        current = self.root
        for node in nodes[1:]:
            current.right = node
            current = node

    def display_tree():
        print("Select a tree: ")
        print("1. Full Binary Tree")
        print("2. Complete Binary Tree")
        print("3. Degenerate Binary Tree")
        print("4. Skewed Binary Tree")
        print("5. Exit")
        tree = input("Select a tree: ")
```
